```python
import pandas as pd
import numpy as np
```

```python
#read the csv file
df=pd.read_csv("accepted_2007_to_2018Q4.csv")
```

```python
#removing all the loans whose status is either not clearly specified. ie(late and current)
keep=['Fully Paid', 'Charged Off','Default',
       'Does not meet the credit policy. Status:Fully Paid',
       'Does not meet the credit policy. Status:Charged Off']
df=df[df['loan_status'].isin(keep)]
#df.shape=(1348099, 151)
```

```python
#dropped all non essential categorical columns
# also dropping funded_amnt_inv 
column=['id',
        'member_id',
        'funded_amnt_inv',
        'grade',
        'emp_title',
        'url',
        'desc',
        'title',
        'pymnt_plan',
        'out_prncp',
        'out_prncp',
        'out_prncp_inv',
        'total_pymnt',
        'total_pymnt_inv',
        'total_rec_prncp',
        'total_rec_int',
        'total_rec_late_fee',
        'recoveries',
        'collection_recovery_fee',
        'last_pymnt_d',
        'last_pymnt_amnt',
        'last_credit_pull_d',
        'last_fico_range_high',
        'last_fico_range_low',
        'collections_12_mths_ex_med',
        'policy_code',
        'hardship_flag',
        'debt_settlement_flag',
        'disbursement_method'
        
        
]
df.drop(columns=column,inplace=True)
# df.shape=(1348099,143)
# dropping all the columns who has more than 90% missing values
p=[]
for cols in df.columns:
    k=df[cols].isnull().mean()*100
    if k>90:
        p.append(cols)
df.drop(columns=p,inplace=True)
# df.shape=(1348099,106)
```

```python
# extracted the columns wise info to do analysis for them
p=pd.DataFrame(df.isnull().mean()*100)
q=pd.DataFrame(df.nunique())
r=pd.DataFrame(df.dtypes)
s=pd.concat([p,q,r],axis=1)
s.to_csv('final_columns.csv')
```

```python
#encoding loan_status to 0 and 1 according the 
#paid,=1
#charged,default=0"
def exchange(s):
    if s in ["Fully Paid",'Does not meet the credit policy. Status:Fully Paid']:
        return 0
    else:
        return 1
df['loan_status']=df['loan_status'].apply(exchange)
```

```python
df.to_csv('cleaned_file.csv')
```
